```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

fontsize=8

plt.rcParams["font.family"] = "Arial"
plt.rcParams["font.size"] = fontsize


fig = plt.gcf()

sizes = []

columns = ['size', 'throughput']

# ssd, 24 workers, 8 v100
df = pd.read_csv('data.dat', sep='\t', names=columns)

size = df['size'].tolist()
throughput = df['throughput'].tolist()

plt.plot(size, throughput, marker='o', markersize=3, linestyle='-', color='blue', lw=1)


plt.ylim(0,100)
plt.xlim(0,1100)
# Adding labels and title
xticks = [64,256,512,1024]
yticks = [0,20,40,60,80,100]
plt.xticks(xticks, fontsize=fontsize)
plt.yticks(yticks, fontsize=fontsize)
plt.xlabel('File size per read (KB)', fontsize=fontsize)
plt.ylabel('Throughput (MB/s)', fontsize=fontsize)
title = plt.title('(c) HDD RAND RD throughput',fontsize=fontsize, pad=10)
title.set_position((0.32, 0))

fig.set_size_inches(1.44, 1)
fig.set_dpi(100)
# plt.legend(fontsize=fontsize)
plt.savefig('thru.eps',  bbox_inches='tight', transparent=True)
# Displaying the chart
plt.show()
```
